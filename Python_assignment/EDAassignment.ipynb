Task 4: Correlation Analysis
● Create a heatmap of correlation between numerical variables.
● Look for potential predictors of churn, e.g., watch_hours, last_login_days,
monthly_fee.

```python
# Task 4
```

```python
numerical_cols = df.select_dtypes(include='number')

corr_matrix = numerical_cols.corr()

plt.figure(figsize=(10, 7))
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

```python
churn_corr = corr_matrix['churned'].drop('churned').sort_values(ascending=False)
print(churn_corr)
```
